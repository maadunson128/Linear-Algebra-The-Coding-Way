**Overview of the Code:**
The code does the following:

1. Imports the `numpy` library.
2. Defines a function to calculate eigenvalues and eigenvectors of a matrix.
3. Defines a function to check if eigenvectors corresponding to distinct eigenvalues are linearly independent. Here, the distinct Eigen values and their corresponding Eigen Vectors are found.
4. Computes eigenvalues and eigenvectors of a matrix.
5. Transposes the eigenvectors since the Eigen vectors are returned as columns.
6. Prints eigenvalues and eigenvectors.
7. Checks for linear independence of distinct eigenvectors.

Overall, it calculates eigenvalues and eigenvectors, then checks if the eigenvectors are linearly independent.

If the rank (non-zero rows) of the matrix formed by the distinct eigenvectors is equal to the number of unknowns (number of eigenvectors), then it is linearly independent. If not, it is not linearly independent.


```python
# importing libraries
import numpy as np
```

```python
# Defining the function to find Eigen Vectors Eigen values 
def eig_vector_value(matrix):
    # Finding the Eigen values and Eigen Vectors
    eig_values,eig_vectors = np.linalg.eig(matrix)
    return eig_values, eig_vectors
    
```

```python
# Defining a function to linear independence of the Eigen Vectors
def check_linear_independence(eigen_vectors):

    # Finding the distinct Eigen Values and their corresponding Eigen Vectors
    distinct_eig_values = []
    distinct_eig_vectors = []
    
    for i in range(len(eigen_values)):
        if eigen_values[i] not in distinct_eig_values:
            distinct_eig_values.append(eigen_values[i])
            distinct_eig_vectors.append(eigen_vectors[i])

    # Finding the rank of the Eigen vectors combined as a single matrix
    matrix = distinct_eig_vectors
    rref_matrix_rank = np.linalg.matrix_rank(matrix)
    
# Checking for linear independence
    if rref_matrix_rank == len(distinct_eig_vectors):
        print("The Eigen Vectors corresponding to distinct eigen values are linearly independent.")
    else:
        print("The Eigen vectors corresponding to distinct eigen values are linearly dependent")
        
```

```python
# Matrix is stored in A in a list
A = [[2, 1, 0], [1, 3, -1], [0, -1, 4]]
eigen_values , eigen_vectors = eig_vector_value(A)
# Eigen vectors are returned as colums, they are transposed to eigen vectors to be in rows
eigen_vectors = eigen_vectors.transpose()
```

```python
print("Eigen Values:",eigen_values,"\n","Eigen Vectors:",eigen_vectors,sep = '\n')
```

```python
# Checking for linear independence
check_linear_independence(eigen_vectors)
```
