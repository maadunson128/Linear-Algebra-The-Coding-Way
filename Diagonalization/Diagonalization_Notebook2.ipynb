```python
# Importing libraries 
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Orginal matrix A
A = np.array([
    [4, 2],
    [1, 5]
])
```

```python
# P matrix and its inverse
P = np.array([
    [2, 1],
    [-1, 1]
])

P_inverse = np.array([
            [1/3, -1/3],
            [1/3, 2/3]
        ])
```

**Here, the `P_inverse` matrix will help us to change any vector from standard coordinates to b coordinates.
Like `P_inverse` matrix, The `P` matrix helps us to change any vector from changed basis coordinates to standard coordinates.**

```python
# Diagonal matrix for A
D = np.array([
    [3, 0],
    [0, 6]
])
```

**The diagonal matrix helps us to easily stretch(scale) the matrix in b coordinates.**

**P is invertible. Therefore the columns of P
are linearly independent. Why?**

P is invertible means that we can find the inverse of the matrix using Gauss elimination method in where the columns have to be linearly independent. If the columns are linearly dependent, eventually the columns will become as zero column, and we can't further proceed the Gauss elimination method.

```python
# New basis vectors in R^2 space (co-ordinates of changed basis )
b1 = np.array([2, -1])
b2 = np.array([1, 1])
```

**B is a new basis in R^2. Why?**

We know that the properties of the basis vector.
- The set of vectors is said to basis vectors of R^n space if the vectors in the set are linearly independent.
- The set of vectors has to span all possible vectors in R^n space.

```python
# X vector
x = np.array([2, 1])
```

The 'x' vector is going to be transformed by a matrix. Here, we are going to do the same but using the diagonalisation method with the help of a change of basis.

From the slides, we know that if we multiply the change of coordinate axes matrix with the x vector, it will give the x vector in b coordinates.

```python
# Finding the x vector in b coordinates 
x_b_coord = np.dot(P_inverse, x)
print(f"x vector on b coordinates:{x_b_coord}")
```

```python
#strecting the vector x on b coordinates
x_transformed_b = np.dot(D, x_b_coord)
print(f"Streched (scaled) x vector on b coordinates :{x_transformed_b} ")
```

**Here, the Diagonal matrix `D` helps us to scale the x vector on co-ordinates b**

```python
#changing the transformed x vector from b co-ordinates to standard co-ordinates
x_transformed_std = np.dot(P, x_transformed_b)
print(f"Transformed x vector in standard co-ordinates : {x_transformed_std}")
```

**The P matrix used to change the transformed vector in b co-ordinates to standard co-ordinates by matrix multiplication**
