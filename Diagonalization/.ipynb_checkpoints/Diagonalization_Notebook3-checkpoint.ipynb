**By Diagonalisation theorem, The eigen vectors of A are the columns of the matrix P and the eigen vectors has to be linearly independent.The diagonal entries of the Diagonal matrix D will be the Eigen values of A that correspond, respectively, to the eigen vectors in P.** 

```python
#importing libraries
import numpy as np
```

```python
#Matrix A
A = [[1, 3, 3],
    [-3, -5, -3],
    [3, 3, 1]]
```

```python
#defining function to calculate eigen values and eigen vectors
def eigen_values_vectors(matrix):
    eig_val, eig_vector = np.linalg.eig(matrix)
    return eig_val, eig_vector
```

```python
#Finding P (Eigen vectors of A)
eigen_values, eigen_vectors = eigen_values_vectors(A)
P = eigen_vectors
print("P matrix:", P, sep= '\n')
P = P.tolist()
```

```python
#Finding diagonal matrix D
D = np.diag(eigen_values)
D = D.tolist()
print(f"Diagonal matrix D : \n {D}")
```

```python
#Defining a function to find the inverse
def inverse(matrix):
    aug_matrix = [row[:] for row in matrix]
    row_length = len(aug_matrix)
    column_length = len(aug_matrix[0])

    #Adding augmented matrix
    for row in range(row_length):
        for column in range(column_length):
            if row == column:
                aug_matrix[row].append(1)
            else:
                aug_matrix[row].append(0)
    
    for j in range(column_length):
        pivot_row = None
        # Finding pivot row 
        for i in range(j, row_length):
            if aug_matrix[i][j] != 0:
                pivot_row = i
                break
        if pivot_row == None:
            raise ValueError("The matrix is singular. Inverse can't be found")

        # Interchange the rows with the pivot row.This helps if the first element is zero in the column.
        aug_matrix[j], aug_matrix[pivot_row] = aug_matrix[pivot_row], aug_matrix[j]

        # Dividing the pivot row by its diagonal element to make the diagonal element 1.
        pivot_value = aug_matrix[j][j]
        for k in range(len(aug_matrix[0])):
            aug_matrix[j][k] = aug_matrix[j][k] / pivot_value

        #Making the non-pivot elements zero
        for i in range(row_length):
            if i != j:
                multiplier = aug_matrix[i][j]
                for k in range(len(aug_matrix[i])):
                    aug_matrix[i][k] -= multiplier * aug_matrix[j][k]

    inverse_matrix = [row[column_length:] for row in aug_matrix]
    return inverse_matrix
                    
                
        
```

```python
# Finding the inverse of matrix P
P_inverse = inverse(P)
print("P Inverse matrix:")
for row in P_inverse:
    print(row)
```

```python
# Defining function to perfrom matrix multiplication
def matrix_multiplication(matrix_1, matrix_2):
    # Checking for the same number of columns and rows
    if len(matrix_1[0]) != len(matrix_2):
        raise ValueError("Matrix multiplication can't be done due to inequal dimensions of rows and columns.")

    # Iterating through each row of matrix_1
    result = []
    for row in range(len(matrix_1)):
        # Iterating through each column of matrix_2
        temp = []
        for column in range(len(matrix_2[0])):
            # Iterating through elements along the respective row and column
            sum = 0
            for k in range(len(matrix_1)):
                sum += matrix_1[row][k] * matrix_2[k][column]
            temp.append(sum)
        result.append(temp)
    return result
    
```

```python
#Performing the multiplication of PDP_inv
product_of_DP_inv = matrix_multiplication(D, P_inverse)
product_of_PDP_inv = matrix_multiplication(P, product_of_DP_inv)
#printing the result
for row in product_of_PDP_inv:
    print(row)
```
