**What does T means ?**

Here, `T` is a linear transformation done on input vector `x`. The transformation is done by matrix multiplication of diagonal matrix `D` with the vector `x` which results in the transformed vector.

```python
# Importing libraries
import matplotlib.pyplot as plt
```

```python
# Basis vectors e1 and e2 
e1 = [1, 0]
e2 = [0, 1]
```

```python
# defining function to transform the input vector 
def transform(vector):
    diagonal_matrix = [[3, 0],
                      [0, 6],
                      ]

    diagonal_matrix_rows = len(diagonal_matrix)
    diagonal_matrix_columns = len(diagonal_matrix[0])
    vector_rows = len(vector)
    
    if diagonal_matrix_columns != vector_rows:
        print("Transformation can't be done.")
        return

    result = []
    # Iterating through diagonal matrix rows
    for i in range(diagonal_matrix_rows):
        sum = 0
        # Iterating through diagonal matrix column elemnts for row i
        for j in range(diagonal_matrix_columns):
            sum += diagonal_matrix[i][j] * vector[j]

        result.append(sum)
    return result
    
```

```python
# Diagonal matrix D
D = [
    [3, 0],
    [0, 6]
]
```

```python
# Finding D^2 matrix (D x D)
result = []
# Iterating through rows
for i in range(len(D)):
    temp_row = []
    # Iterating through columns
    for j in range(len(D[0])):
        sum = 0
        # Iterating through the elements of the diagonal matrices to compute the product and sum
        for k in range(len(D)):
            sum += D[i][k] * D[k][j]
        temp_row.append(sum)
    result.append(temp_row)
```

```python
# printing D^2 matrix
for element in result:
    print(element)
```

```python
# Matrix A with some random elemnts
A = [[1, 2],
    [3, 4]]
```

```python
# Finding D x A matrix
result = []
# Iterating through rows  
for i in range(len(D)):
    temp_row = []
    # Iterating through columns
    for j in range(len(D[0])):
        sum = 0
        # Iterating through the elements of the diagonal matrix D and A to compute the product and sum
        for k in range(len(D)):
            sum += D[i][k] * A[k][j]
        temp_row.append(sum)
    result.append(temp_row)
```

```python
# Printing the transformed matrix
for element in result:
    print(element)
```

**D is called as Diagonal matrix. Calculate D^2 to understand why diagonal matrices are useful.**
From both results, we can say that diagonal elements are the scaling factor in the diagonal matrix.The each row in the matrix gets scaled by the diagonal elements of the respective diagonal element.

By using diagonal matrix, we can scale the axes (x axis, y axis) independely without affecting the other axes.
 

```python
# Transforing the basis vector e1 and e2
trans_e1 = transform(e1)
trans_e2 = transform(e2)

print("Transformed Basis vector e1:",trans_e1,sep = '\n')

print("Transformed Basis vector e2:",trans_e2,sep = '\n')
```

```python
plt.figure(figure = (6,6))
# Plotting the e1,e2 , transformed e1 and e2
plt.plot([0, e1[0]], [0, e1[1]], color='r', label='e1')
plt.plot([0, e2[0]], [0, e2[1]], color='m', label='e2')
plt.plot([0, trans_e1[0]], [0, trans_e1[1]], color='c', label='trans_e1')
plt.plot([0, trans_e2[0]], [0, trans_e2[1]], color='y', label='trans_e2')

# Adding arrows to the vectors
plt.arrow(0, 0, e1[0], e1[1], color='r', head_width = 0.15, head_length = 0.15)
plt.arrow(0, 0, e2[0], e2[1], color='m', head_width = 0.15, head_length = 0.15)
plt.arrow(0, 0, trans_e1[0], trans_e1[1],color = 'c', head_width = 0.15, head_length = 0.15)
plt.arrow(0, 0, trans_e2[0], trans_e2[1], color = 'y', head_width = 0.15, head_length = 0.15)

# Adding text above the arrow mark
plt.text(e1[0] + 0.1, e1[1] + 0.1, 'e1 [1, 0]', fontsize=10, color='r')
plt.text(e2[0] + 0.1, e2[1] + 0.1, 'e2 [0, 1]', fontsize=10, color='m')
plt.text(trans_e1[0] + 0.1, trans_e1[1] + 0.1, 'trans_e1 [3, 0]', fontsize=10, color= 'c')
plt.text(trans_e2[0] + 0.1, trans_e2[1] + 0.1, 'trans_e2 [0, 6]', fontsize=10, color='y')


# Setting x axis and y axis limit
plt.xlim(-0.5,4)
plt.ylim(-0.5,8)

#Plotting the vectors with grid and legend 
plt.grid()
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

***From the plot, We can see that diagonal matrix transformed the basis vectors e1, e2 with diagonal elements as scaling factors.***

```python
# Transforming matrix x
x = [1, 1]
trans_x = transform(x)

print("Transformed x vector:",trans_x ,sep = '\n')
```

```python
# Plotting the vector x and transformed vector x
plt.figure(figure = (6,6))
plt.plot([0, x[0]], [0, x[1]], color = 'r', label = 'x')
plt.plot([0, trans_x[0]], [0, trans_x[1]], color = 'b', label = 'trans_x')

# Adding arrows for the vectors x and trans_x
plt.arrow(0, 0, x[0], x[1], color = 'r', head_width = 0.1, head_length = 0.15)
plt.arrow(0, 0, trans_x[0], trans_x[1], color = 'b', head_width = 0.1, head_length = 0.15)

# Adding text above the arrow for x and trans_x vector
plt.text(x[0] + 0.15, x[1] + 0.15, 'x [1, 1]', fontsize = 12)
plt.text(trans_x[0] + 0.15, trans_x[1] + 0.15, 'trans_x [3,6]', fontsize = 12)

# Setting x axis and y axis limit
plt.xlim(0,6)
plt.ylim(0,10)

#Plotting the vectors with grid and legend 
plt.grid()
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

***From the plot, It is clearly tell that the diagonal elements scaled up the respective row elements in the vector.***
